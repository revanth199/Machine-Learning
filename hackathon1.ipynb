# KNN

```python
 # x=a1[["fixed acidity","volatile acidity","citric acid",'total sulfur dioxide','density','sulphates','alcohol','chlorides','free sulfur dioxide','residual sugar','pH']]
```

```python
 x=a1.iloc[:,0:11]
```

```python
x
```

```python
y = a1.iloc[:,11]
```

```python
y
```

```python
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
x_train
```

```python
y_train=y_train.astype('int')
y_train
```

```python
x_test
```

```python
y_test=y_test.astype('int')
y_test
```

```python
#Fitting classifier to training set
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train,y_train)
```

```python
#predict on test data
y_pred=clf.predict(x_test)
```

```python
print('Actual quality')
print(y_test.values)
```

```python
print('Predicted quality')
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("\nAccuracy score:%f"%(accuracy_score(y_test,y_pred)*100))
print("ROC score:%f\n"%(roc_auc_score(y_test,y_pred)*100))  # Receiver operating characteristics(ROC)
print("\nRecall score:%f"%(recall_score(y_test,y_pred)*100)) 
print(confusion_matrix(y_test,y_pred))
```

```python
#evaluation metrics
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
# print cross validation score for KNN

clf_acc = cross_val_score(clf,x_train_std,y_train,cv=3,scoring= "accuracy",n_jobs = -1)
```

```python
clf_acc
```

```python
clf_proba = cross_val_predict(clf,x_train_std,y_train,cv=3,method="predict_proba")
```

```python
clf_scores = clf_proba[:,1]
```

```python
clf_scores
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
def ROC_curve(title,y_train,scores,label=None):
    
      # calculating the ROC_score
    fpr,tpr,thresholds = roc_curve(y_train,scores)
    print('AUC Score({}):{:2f}'.format(title,roc_auc_score(y_train,scores)))
    plt.figure(figsize=(8,6))
    plt.plot(fpr,tpr,linewidth=2,label=label,color='b')
    plt.xlabel('False Positive Rate',fontsize = 16)
    plt.ylabel('True positive Rate',fontsize = 16)
    plt.title('ROC Curve:{}'.format(title),fontsize=16)
    plt.show()    
    
```

```python
ROC_curve('kNN',y_train,clf_scores)
```
